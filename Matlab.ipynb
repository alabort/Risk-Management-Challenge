```python
H=[]

for x in Altitude:
    while True:
        try:
            H.append(int(x))
            break
        except TypeError:
            H.append(0)
            break
```

```python
import numpy as np
```

```python
# Monte carlo simulation - modelled as a function of damage
exp_dam = np.random.normal(0, 1, size=  (29,10)) # example with normal distribution with mu=0 and sigma=1
# until 2050 - 28 years
print(exp_dam)
```

```python
# we prepare the array where we put the expected damages
futureDamage=np.zeros((29,10))
# we put initial damage as 0
futureDamage[0]=0
drift=0.2
```

```python
for i in range(1,29):
    futureDamage[i]=(exp_dam[i]+drift)*758.122199+futureDamage[i-1]
print(futureDamage)
```

```python
plt.rcParams["figure.figsize"] = [20,10]
plt.plot(futureDamage)
plt.show()
```
